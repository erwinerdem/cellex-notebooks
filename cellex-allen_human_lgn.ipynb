# CELLEX

```python
import pandas as pd
import cellex
```

```python
data_dir = 'data/CELLEX'
file_s = data_dir+'/Allen_Human_Lateral_geniculate/human_LGN_2018-06-14_samples-columns.csv'
file_g = data_dir+'/Allen_Human_Lateral_geniculate/human_LGN_2018-06-14_genes-rows.csv'
file_e = data_dir+'/Allen_Human_Lateral_geniculate/human_LGN_2018-06-14_exon-matrix.csv'
file_i = data_dir+'/Allen_Human_Lateral_geniculate/human_LGN_2018-06-14_intron-matrix.csv'
```

```python
%%time
samples = pd.read_csv(file_s)
genes = pd.read_csv(file_g)
exon = pd.read_csv(file_e)
intron = pd.read_csv(file_i)

exon.drop('Unnamed: 0',axis=1, inplace=True)
intron.drop('Unnamed: 0',axis=1, inplace=True)
```

```python
display(genes)
```

```python
data = exon+intron
data.index = genes.gene
display(data)
```

```python
metadata = samples[['sample_name','cluster']]
display(metadata)# cluster name cannot contain spaces, other CELLECT will return an error
```

```python
metadata['cluster'].replace(' ', '_', regex=True, inplace=True)
metadata.set_index('sample_name', inplace=True)
display(metadata)
```

```python
metadata.cluster.unique() # contains the cluster no_class
```

```python
nan_cells = metadata[metadata['cluster']=='no_class'].index
len(nan_cells)
```

```python
data_dropped = data.drop(nan_cells, axis=1)
print(data.shape, data_dropped.shape)
```

```python
metadata_dropped = metadata[metadata['cluster']!='no_class']
print(metadata.shape, metadata_dropped.shape)
```

```python
%%time
eso = cellex.ESObject(data=data_dropped,
                      annotation=metadata_dropped,
                      verbose=True)
```

```python
%%time
eso.compute(verbose=True)
```

```python
eso.results['esmu']
```

```python
cellex.utils.mapping.human_symbol_to_human_ens(eso.results['esmu'], drop_unmapped=True, verbose=True)
```
