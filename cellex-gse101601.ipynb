# CELLEX

```python
import pandas as pd
import cellex
```

```python
data_dir = 'data/CELLEX/'
file = data_dir+'/GSE101601_Temporal_Cortex/GSE101601_molecules_human.csv'
```

```python
%%time
data = pd.read_csv(file, 
                   index_col=0,
                   sep=';')
```

```python
data
```

```python
metadata = pd.DataFrame()
metadata['cluster'] = data.loc['cluster']
metadata
```

```python
data_dropped = data[1:]
display(data_dropped)
```

```python
print(data_dropped.shape, metadata.shape)
# (28274, 2028) (2028, 1)
```

```python
%%time 
eso = cellex.ESObject(data=data_dropped,
                      annotation=metadata,
                      verbose=True)
```

```python
%%time 
eso.compute(verbose=True)
```

```python
eso.results['esmu']
```

```python
cellex.utils.mapping.human_symbol_to_human_ens(eso.results["esmu"], drop_unmapped=True, verbose=True)
```
