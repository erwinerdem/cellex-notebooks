# CELLEX

```python
import pandas as pd
import cellex
```

```python
data_dir = 'data/CELLEX/'
file = data_dir+'/GSE67835_Human_Cortex/GSE67835_count.txt'
```

```python
%%time 
data = pd.read_csv(file, 
                   index_col=0,
                   sep='\t')
```

```python
data.index # contains whitespace
```

```python
data.index = data.index.str.strip() #remove whitespace after gene name
```

```python
data
```

```python
# remove last 3 rows
data_dropped = data[:-3]
display(data_dropped)
```

```python
file = data_dir+'/GSE67835_Human_Cortex/GSE67835_metadata.txt'
metadata = pd.read_csv(file, sep='\t',header=0)
metadata = metadata.set_index('cell_id').drop('age',axis=1)
display(metadata)
```

```python
print(data.shape,metadata.shape)
```

```python
%%time 
eso = cellex.ESObject(data=data_dropped, 
                      annotation=metadata,
                      verbose=True)
# Preprocessing - checking input ... input parsed in 0 min 0 sec
# Preprocessing - running remove_non_expressed ... excluded 458 / 22088 genes in 0 min 0 sec
# Preprocessing - normalizing data ... data normalized in 0 min 0 sec
# Preprocessing - running ANOVA ... excluded 18167 / 21630 genes in 0 min 3 sec
# eso time: 4.198008060455322
```

```python
%%time 
eso.compute(verbose=True)
```

```python
cellex.utils.mapping.human_symbol_to_human_ens(eso.results["esmu"], drop_unmapped=True, verbose=True)
eso.results['esmu']
```

```python
eso.results["esmu"].to_csv("GSE67835_Human_Cortex.esmu.csv.gz")
```
